```python
import os
import random
import time
import numpy as np
import pandas as pd
import torch
from faker import Faker
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments

# 1. WandB Girişini Otomatik Kapat (Sürekli sormasın)
os.environ["WANDB_DISABLED"] = "true"

# 2. Tekrarlanabilirlik Ayarları
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🚀 Çalışma Ortamı Hazır! Kullanılan Cihaz: {device}")

# ==========================================
# AŞAMA 1: %100 TÜRKÇE VE MANTIKLI VERİ ÜRETİMİ
# ==========================================
def generate_turkish_business_data(num_rows=50000):
    print(f"📊 {num_rows} satırlık %100 TÜRKÇE ve MANTIKLI veri seti üretiliyor...")

    departments = ['Lojistik', 'İK', 'Finans', 'Üretim', 'IT', 'Satış', 'Müşteri Hizmetleri']

    # Giriş Kalıpları
    prefixes = [
        "Sistem uyarısı:", "Vardiya raporu:", "Otomatik bildirim:",
        "Kullanıcı notu:", "Günlük akış:", "Denetim kaydı:", "Acil durum:", ""
    ]

    # Darboğaz Sebepleri (Sorunlar)
    bottleneck_reasons = [
        "sunucu yanıt vermiyor", "stok seviyesi kritik eşiğin altında",
        "onay sürecinde tıkanıklık var", "veritabanı kilitlendi (deadlock)",
        "personel eksikliği nedeniyle hat durdu", "sevkiyat aracı arızalandı",
        "yazıcı kuyruğu sıkıştı", "aşırı ısınma tespit edildi",
        "hatalı hammadde girişi reddedildi", "fatura sistemi cevap vermiyor"
    ]

    # Normal Süreçler (İşler yolunda)
    normal_activities = [
        "periyodik bakım tamamlandı", "sistem yeniden başlatıldı",
        "stok sayımı sorunsuz bitti", "vardiya değişimi yapıldı",
        "test ortamı stabil çalışıyor", "fatura başarıyla onaylandı",
        "paketleme işlemi standartlara uygun", "veri yedekleme tamamlandı",
        "güvenlik taraması temiz çıktı", "yeni sipariş sisteme girildi"
    ]

    # Latince Yerine Gelecek EK NOTLAR (Tamamen Türkçe)
    extra_notes = [
        "İlgili birim bilgilendirildi.", "Yönetici onayı bekleniyor.",
        "Detaylar ekli dosyada mevcuttur.", "Müşteri ile teyitleşildi.",
        "Süreç prosedüre uygun ilerliyor.", "Teknik ekip yönlendirildi.",
        "Kayıtlar arşivlendi.", "Lütfen manuel kontrol sağlayın.",
        "Bir sonraki vardiyaya not bırakıldı.", "İşlem logları temizlendi."
    ]

    data = []

    for i in range(num_rows):
        dept = random.choice(departments)
        prefix = random.choice(prefixes)
        extra = random.choice(extra_notes)

        # %40 Darboğaz, %60 Normal (Dengesiz veri simülasyonu)
        is_bottleneck = 1 if np.random.random() < 0.4 else 0

        if is_bottleneck:
            reason = random.choice(bottleneck_reasons)
            sentence = f"{prefix} {dept} bölümünde {reason}. {extra}"
            duration = np.random.randint(45, 600)
        else:
            activity = random.choice(normal_activities)
            sentence = f"{prefix} {dept}: {activity}. {extra}"
            duration = np.random.randint(5, 45)

            # Doğallık için bazen küçük harfe çevir
            if np.random.random() < 0.3:
                sentence = sentence.lower()

        data.append([dept, sentence, duration, is_bottleneck])

    random.shuffle(data)
    df = pd.DataFrame(data, columns=['Department', 'Log_Text', 'Duration_Min', 'Is_Bottleneck'])
    df.to_csv('business_process_data.csv', index=False)
    print("✅ Veri üretildi: Latince kelimeler temizlendi, tamamen Türkçe.")
    return df

# Veriyi üret
df = generate_turkish_business_data(50000)

# ==========================================
# AŞAMA 2: MODEL GELİŞTİRME VE EĞİTİM
# ==========================================

def clean_text(text):
    return text.lower().strip()

print("🛠️ Veri ön işleme ve split işlemi yapılıyor...")
df['cleaned_text'] = df['Log_Text'].apply(clean_text)

X = df['cleaned_text'].values
y = df['Is_Bottleneck'].values

# Split
split = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=42)
for train_index, temp_index in split.split(X, y):
    X_train, X_temp = X[train_index], X[temp_index]
    y_train, y_temp = y[train_index], y[temp_index]

split2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
for val_index, test_index in split2.split(X_temp, y_temp):
    X_val, X_test = X_temp[val_index], X_temp[test_index]
    y_val, y_test = y_temp[val_index], y_temp[test_index]

print(f"Eğitim: {len(X_train)}, Doğrulama: {len(X_val)}, Test: {len(X_test)}")

# Tokenization
print("🔄 Tokenization işlemi başlatılıyor (DistilBert)...")
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-multilingual-cased')

class BusinessDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    def __len__(self):
        return len(self.labels)

def tokenize_data(texts):
    return tokenizer(texts.tolist(), truncation=True, padding=True, max_length=64)

train_dataset = BusinessDataset(tokenize_data(X_train), y_train)
val_dataset = BusinessDataset(tokenize_data(X_val), y_val)
test_dataset = BusinessDataset(tokenize_data(X_test), y_test)

# Metrikler
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    return {'accuracy': acc, 'f1': f1, 'precision': precision, 'recall': recall}

# EĞİTİM (Düzeltilmiş Parametreler)
print("🧠 Model eğitimi başlıyor... (GPU ile yaklaşık 5-8 dk)")
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-multilingual-cased', num_labels=2)
model.to(device)

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=100,

    # GÜNCEL PARAMETRELER (Hata vermez)
    eval_strategy="steps",
    eval_steps=500,
    save_strategy="steps",
    save_steps=500,
    load_best_model_at_end=True,
    report_to=None # WandB kapalı
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

trainer.train()

# Kaydet
model_path = "./saved_model"
model.save_pretrained(model_path)
tokenizer.save_pretrained(model_path)
print(f"✅ Model başarıyla eğitildi ve '{model_path}' klasörüne kaydedildi.")

# ==========================================
# AŞAMA 3: GÜNCEL STREAMLIT DOSYASI (app.py)
# ==========================================
# Gelişmiş arayüz kodunu buraya gömüyoruz.

app_code = """
import streamlit as st
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
import torch
import torch.nn.functional as F
import pandas as pd
import os

st.set_page_config(page_title="Business AI", layout="wide", page_icon="🏭")

@st.cache_resource
def load_model():
    if not os.path.exists("./saved_model"): return None, None
    tokenizer = DistilBertTokenizer.from_pretrained("./saved_model")
    model = DistilBertForSequenceClassification.from_pretrained("./saved_model")
    return tokenizer, model

tokenizer, model = load_model()

st.title("🏭 Kurumsal Süreç Darboğaz Tespit Sistemi")

if model is None:
    st.error("Model bulunamadı. Lütfen önce eğitimi tamamlayın.")
    st.stop()

col1, col2 = st.columns([1, 1])

with col1:
    st.subheader("📝 Log Girişi")
    user_input = st.text_area("Log Metni:", placeholder="Örn: Sevkiyat aracı arızalandı, teslimat gecikecek.", height=150)
    analyze_btn = st.button("🔍 Analiz Et", type="primary", use_container_width=True)

    st.info("Hızlı Test:")
    if st.button("Örnek: Finans Sorunu"): st.code("Finans onayı bütçe aşımı nedeniyle durdu.")

def get_suggestion(text):
    text = text.lower()
    if "sevkiyat" in text or "araç" in text: return "🚚 **Lojistik:** Alternatif rota veya kargo firması devreye alınmalı."
    if "sunucu" in text or "sistem" in text: return "💻 **IT:** Sistem yükü kontrol edilmeli, IT ekibine ticket açılmalı."
    if "fatura" in text or "finans" in text: return "💰 **Finans:** Bütçe onayı için yönetici eskalasyonu yapılmalı."
    if "üretim" in text or "makine" in text: return "⚙️ **Üretim:** Bakım ekibi hatta yönlendirilmeli."
    return "📋 **Genel:** Süreç yöneticisi bilgilendirilmeli."

with col2:
    st.subheader("📊 Sonuç")
    if analyze_btn and user_input:
        with st.spinner('Analiz ediliyor...'):
            inputs = tokenizer(user_input, return_tensors="pt", truncation=True, max_length=64)
            with torch.no_grad():
                probs = F.softmax(model(**inputs).logits, dim=1)

            bottleneck_prob = probs[0][1].item()
            if bottleneck_prob > 0.5:
                st.error(f"🚨 DARBOĞAZ TESPİT EDİLDİ (%{bottleneck_prob*100:.1f})")
                st.progress(bottleneck_prob)
                st.warning(get_suggestion(user_input))
            else:
                st.success(f"✅ SÜREÇ NORMAL (%{probs[0][0].item()*100:.1f})")
                st.progress(probs[0][0].item())

st.sidebar.header("Canlı Veri (Simülasyon)")
if st.sidebar.button("Veri Getir"):
    try:
        st.sidebar.dataframe(pd.read_csv('business_process_data.csv').sample(5)[['Log_Text', 'Is_Bottleneck']], hide_index=True)
    except: st.sidebar.error("Veri yok.")
"""

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("✅ Streamlit dosyası (app.py) güncellendi. Cloudflare ile başlatmaya hazırsınız.")
```
